**We now use the properties in Box 4.1 to derive the response distribution. a) Combine the first two properties to show that the random variable aX + bY has mean aμX + bμY and variance a2σ 2 X + b2σ 2 Y. b) Using this result, prove Eqs. (4.3) and (4.4).**  


**A**
$$aX + bY$$
We know that the mean of of the random variable $X + Y$ is $\mu_X + \mu_Y$, and the variance is $\sigma_X^2 + \sigma_Y^2$. Additionally, the mean of a random variable $aX$, where $a$ is a constant, is $a\mu$, with a variance of $a^2\mu^2$. From this we can get the following equation for the mean of the random variable $aX + bY$ is $a\mu_X + b\mu_Y$. The variance is then $a^2\sigma_X^2 + b^2\sigma_Y^2$.


**B**  
From the textbook we know the folowing.  
Equation 4.3: 
$$\mathbb{E}[\hat{s}_{PM}|s] = \frac{Js+J_s\mu}{J+J_s}$$
Equation 4.4:
$$Var[\hat{s}_{PM}|s] = \frac{J}{(J+J_s)^2}$$
Where:
$$J = \frac{1}{\sigma^2}\; \textrm{and}\;J_s = \frac{1}{\sigma_s^2}$$
   

$S_{PM}$ is a linear function of $x_{obs}$, and when the stimulus is $s$, $x_{obs}$ has a mean of $s$ and a variance of $\sigma^2$.   
From equation 4.1 we can make:
$$\mu_{post}=ax_{obs} + b$$
where  
$$a=\frac{J}{J+J_s} \; \textrm{and}\; b=\frac{J_s}{J+J_s}\mu$$   
Since $x_{obs}$ has a mean of $s$, the mean of $S_{PM}$ (since it is a linear function of $x_{obs}$) is $as + b$. So we get the following:
$$\mathbb{E}[\hat{s}_{PM}|s] = as + b$$
and if we replace a and b with thier full representations we get equation 4.3:
$$\mathbb{E}[\hat{s}_{PM}|s] = \frac{Js+J_s\mu}{J+J_s}$$  
From this we can also prrove equation 4.4:  
The variance of a random variable $aX + b$ is $a^2\sigma^2$, and we get:
$$Var[\hat{s}_{PM}|s]= a^2\sigma^2$$
or
$$Var[\hat{s}_{PM}|s] = (\frac{J}{J+J_s}^2)\sigma^2$$
which leads to
$$Var[\hat{s}_{PM}|s] = (\frac{J^2}{(J+J_s)^2})(\frac{1}{J})$$
leading to our final equation
$$Var[\hat{s}_{PM}|s] = \frac{J}{(J+J_s)^2}$$




### 4.2
**Why does it make intuitive sense that the standard deviation of the posterior mean estimate decreases again?**   
As the standard deviation of he  of the likelihood continues to increase, our posterior estimate will become more biased by our prior. Since our standard deviation of our response distribution is the variation of of the mean of our posteriors, as the posterior becomes more biased by the prior (which does not vary), this distribution of means will become less variable. Eventually, our posterior will essentially equal our prior, and the response distribution standard devaition will reach zero.

```python
def gaussian_pdf(y, mu, sigma):
    const = 1 / (np.sqrt(2 * np.pi * (sigma**2)))
    exp = np.exp(-(((y - mu) ** 2) / (2 * (sigma ** 2))))
    return const * exp
```

```python
measurement_mean = 10
p_mean = 20
p_var = 4
measurment_var = 5

s = np.arange(-10, 35.2, 0.2) 
prior = gaussian_pdf(s, p_mean, p_var)
x_obs = np.random.normal(loc=measurement_mean, scale=measurment_var, size=5)

print(f"the random values drawn were {x_obs}")


# This could all be done in the loop, but this way the legend only uses the first three lines
fig, ax = plt.subplots(figsize=(10,6))
ax.plot(s, prior, color='red', label="Prior $p(s)$")
likelihood = [gaussian_pdf(x_obs[0], x, measurment_var) for x in s]
protoposterior = prior * likelihood
posterior = protoposterior / np.sum(protoposterior)
posterior /= 0.2
ax.plot(s, likelihood, color='blue', label="Likelihood $p(x|s)$")
ax.plot(s, posterior, color='purple', label="Posterior $p(s|x)$")
ax.set_xlabel('stimulus values')
ax.set_ylabel('probability density')
ax.set_title('Prior')
ax.legend(frameon=False)
sns.despine()

for obs in range(1, len(x_obs)):
    likelihood = [gaussian_pdf(x_obs[obs], x, measurment_var) for x in s]

    # Step C
    protoposterior = prior * likelihood
    # Step F
    posterior = protoposterior / np.sum(protoposterior)
    # Step E
    posterior /= 0.2
    # Step F
    ax.plot(s, likelihood, color='blue')
    ax.plot(s, posterior, color='purple')
    ax.set_xlabel('stimulus values')
    ax.set_ylabel('probability density')
    ax.set_title('Posterior jumping around')
    sns.despine()
```
