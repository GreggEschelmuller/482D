```python
# Imports and functions
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def gaussian_pdf(y, mu, sigma):
    const = 1 / (np.sqrt(2 * np.pi * (sigma**2)))
    exp = np.exp(-(((y - mu) ** 2) / (2 * (sigma ** 2))))
    return const * exp
```

**a. What are the mean and variance of the posterior? You may start with the equations in Section 5.4.**  

$$\hat{s}_PM = \mu_{post} = \frac{ J_s\mu+J_1x_{obs,1}+J_2x_{obs,2}...+J_Tx_{obs,T}}{J_s+J_1+J_2...+J_T}$$

$$\hat{s}_PM =\mu_{post} = \frac{J_s\mu+\sum{J_tx_{obs,t}}}{J_s + \sum{J_t}}$$

Since each measurement has an identical variance, we know that $J_t=\frac{1}{\sigma^2}$ for all of our $x_{obs}$.
So $\mu_{post}$ can be generalized to the following

$$\mu_{post} = \frac{\sum_{t=1}^{T}J_tx_{obs}}{\sum_{t=1}^{T}J_t}$$
or
$$\mu_{post} = \frac{\sum_{t=1}^{T}J_tx_{obs}}{T*J_t}$$
Further simplified to:
$$\mu_{post} = \frac{J_t\sum_{t=1}^{T}x_{obs}}{T*J_t}$$
And lastly
$$\mu_{post} = \frac{\sum_{t=1}^{T}x_{obs}}{T}$$



Equation 5.28 is:
$$\sigma_{post}^2 = \frac{1}{J_s + \sum_{i=1}^{N} J_i}$$
Which can be simplified to:
$$\sigma_{post}^2 = \frac{1}{\sum_{t=1}^{T}J_t}$$
$$\sigma_{post}^2 = \frac{1}{T*J_t}$$
which can be finally simplified to:
$$\sigma_{post}^2 = \frac{\sigma_s^2}{T}$$

### 5.6
**In this problem, we examine suboptimal estimation in the context of cue combination. Suppose an observer estimates a stimulus s from two conditionally independent, Gaussiandistributed measurements, xobs,1 and xobs,2. The prior is flat.**   

**a. We start with a reminder of optimal estimation. Express the PME in terms of the measurements.**   

$$PME = \frac{J_1x_{obs,1}+J_2x_{obs,2}}{J_1 + J_2}$$

**b. What is the variance of the PME across trials?**

$$\sigma = \frac{1}{J_1 + J_2}$$

**c. Now suppose the observer uses an estimator of the form $\hat{s} = wx_{obs,1} + (1-w)x_{obs,2}$, where w can be any constant. Show that this estimate is unbiased (just like the PME); this means that the mean of the estimate for given s is equal to s .**  

The expected value of $\hat{s}$ is:

$$\mathbb{E}[\hat{s}] = w \mathbb{E}[x_{obs,1}] + (1 - w) \mathbb{E}[x_{\text{obs},2}]$$

Since we are assuming that the observation woud be centered on the true stimulus value, we can get the following:

$$\mathbb{E}[\hat{s}] = w s + (1 - w) s$$

$$\mathbb{E}[\hat{s}] = s(w + 1 - w)$$

$$\mathbb{E}[\hat{s}] = s$$

**d. What is the variance of this estimate as a function of w? Plot this function. At which value of w is it minimal, and does this value make sense? State your final conclusion in words**

$$J_{final} = J_1 + J_2$$

$$w = \frac{J_1}{J_1+J_2}$$

Note: There is a global minimum.

$$var = w^2\sigma_1^2 + (1-w)^2\sigma_2^2$$

```python
w_values = np.arange(0,1,0.01)
fig, ax = plt.subplots()
sigmas_1 = np.array([0.5, 1, 1.5, 2])
sigmas_2 = np.array([1.5, 0.1, 2, 1])
for i in range(len(sigmas_1)):
    s1 = sigmas_1[i]
    s2 = sigmas_2[i]
    w_square = w_values**2
    w_minus = (1-w_values)**2
    variances = w_square*s1 + w_minus*s2
    min_val = np.argmin(variances)

    j1 = 1/s1
    j2 = 1/s2
    w_analytic = j1/(j1+j2)
    ax.plot(w_values, variances)
    print(f"Numerical: {round(w_values[min_val],2)}. Analytical: {round(w_analytic,2)}")
ax.set_ylabel("variance")
ax.set_xlabel("w values")
ax.set_title(f"The minimum variance is at w = {w_values[min_val]}")
plt.show()
```

### 5.7 
**In Chapters 3 and the present chapter, we were able to derive analytical expressions for the posterior distribution and the response distribution. For more complex psychophysical tasks (e.g. later in this book), however, analytical solutions often do not exist but we can still use numerical methods. To gain familiarity with such methods, we will work through the cue combination model in this chapter using numerical methods. We assume that the experimenter introduces a cue conflict between the auditory and the visual stimulus: s1 = 5 and s2 = 10. The standard deviations of the auditory and of the visual noise are σ1 = 2 and σ2 = 1, respectively. We assume a flat prior over s.**

**a. Randomly draw an auditory measurement xobs,1 and a visual measurement xobs,2 from their respective distributions.**

```python
s1 = 5  
s2 = 10  
sigma1 = 2  
sigma2 = 1  

xobs1 = np.random.normal(s1, sigma1)
xobs2 = np.random.normal(s2, sigma2)

xobs1, xobs2
```

**b) Plot the corresponding elementary likelihood functions, p(xobs,1|s) and p(xobs,2|s), in one figure.**

```python
step_size = 0.1
s_values = np.arange(0, 15, step_size)

likelihood_auditory = gaussian_pdf(s_values, xobs1, sigma1)
likelihood_visual = gaussian_pdf(s_values, xobs2, sigma2)

# Plotting
fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(s_values, likelihood_auditory, label='Auditory Likelihood', color='blue')
ax.plot(s_values, likelihood_visual, label='Visual Likelihood', color='green')
ax.set_xlabel('Stimulus Value (s)')
ax.set_ylabel('Likelihood')
ax.set_title('Likelihood Functions for Auditory and Visual Measurements')
ax.legend()
plt.show()
```

**d) Calculate the posterior distribution by numerically normalizing the combined likelihood function. Plot this distribution in the same figure as the likelihood functions.**

```python
protoposterior = likelihood_auditory * likelihood_visual
posterior = protoposterior / np.sum(protoposterior)
posterior /= step_size

fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(s_values, likelihood_auditory, label='Auditory Likelihood', color='blue')
ax.plot(s_values, likelihood_visual, label='Visual Likelihood', color='green')
ax.plot(s_values, posterior, label='Posterior', color='purple')
ax.set_xlabel('Stimulus Value (s)')
ax.set_ylabel('density')
ax.set_title('Likelihood Functions for Auditory and Visual Measurements')
ax.legend()
plt.show()
```

**e) Numerically find the PME of s, i.e., the value of s at which the posterior distribution is maximal**

```python
post_mean = np.sum(posterior * s_values)*step_size
post_mean
```

**f) Compare with the PME of s computed from Eq. (5.14) using the measurements drawn in (a).**

$$\mu_{post} = w_1x_{obs,1} + w_2x_{obs,2}$$

$$w_1 = \frac{J_1}{J_1+J_2}$$
$$w_2 = \frac{J_2}{J_1+J_2}$$

```python
j1 = 1/sigma1**2
j2 = 1/sigma2**2

w1 = j1/(j1+j2)
w2 = j2/(j1+j2)

mu_post = w1*xobs1 + w2*xobs2

post_mean,mu_post
```

**g) In the above, we simulated a single trial and computed the observer’s PME of s, given the noisy measurements on that trial. If an analytical solution does not exist for the distribution of the PMEs, we can repeat the above procedure many times to approximate this distribution. Here, we practice this method even though an analytical solution is available in this case. Draw 100 pairs (xobs,1, xobs,2) and numerically compute the observer’s PME for each pair as in (e).**

```python
pme_means = []

for i in range(1,101):
    xobs1 = np.random.normal(s1, sigma1)
    xobs2 = np.random.normal(s2, sigma2)

    likelihood_auditory = gaussian_pdf(s_values, xobs1, sigma1)
    likelihood_visual = gaussian_pdf(s_values, xobs2, sigma2)

    protoposterior = likelihood_auditory * likelihood_visual
    posterior = protoposterior / np.sum(protoposterior)
    posterior /= step_size

    post_mean = np.sum(posterior * s_values)*step_size
    pme_means.append(post_mean)
np.mean(pme_means)
```

**h) Compute the mean of the PMEs obtained in (g) and compare with the mean estimate predicted using Eq. (5.19).**

```python
pme_mean = w1*s1 + w2*s2
pme_mean
```

**i) Make a histogram of the PME.**

```python
fig, ax = plt.subplots()
sns.histplot(pme_means, ax=ax, kde=True)
sns.despine()
```

**j) Relative (auditory) bias is defined as the mean PME minus the true auditory stimulus, divided by the true visual stimulus minus the true auditory stimulus. Compute relative auditory bias for your estimates.**
$$bias = \frac{PME - s}{\mu - s}$$

```python
biases = []
for mean in pme_means:
    bias = (mean - s1)/(s2 - s1)
    biases.append(bias)
fig, ax = plt.subplots()
sns.histplot(biases, ax=ax, kde=True)
sns.despine()
print(np.mean(biases))
```
